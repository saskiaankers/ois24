```r
install.packages("fields")
library(fields)

z_function <- function(x, y) {
  sin(x^2) * cos(y^2)^2
}

# Création d'une grille de valeurs
x <- seq(-2*pi, 2*pi, length.out = 100)
y <- seq(-2*pi, 2*pi, length.out = 100)

# Définition de la fonction
z <- outer(x,y,FUN = z_function)

# graphique
image.plot(x, y, z, col = heat.colors(20), xlab = "Distance x (m)", ylab = "Distance y (m)",
           main = "Hauteur de l'eau à un instant t après le lancer d'un gros caillou")
```
